```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def explore_data(df):
    """Análise exploratória básica"""

    # Estatísticas descritivas
    print("Estatísticas Descritivas:")
    print(df.describe())

    # Verificar missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Visualizações
    fig, axes = plt.subplots(3, 1, figsize=(15, 10))

    # Preço de fechamento
    axes[0].plot(df.index, df['Close'])
    axes[0].set_title('Preço de Fechamento')

    # Volume
    axes[1].bar(df.index, df['Volume'])
    axes[1].set_title('Volume de Negociação')

    # Retornos diários
    df['Returns'] = df['Close'].pct_change()
    axes[2].plot(df.index, df['Returns'])
    axes[2].set_title('Retornos Diários')

    plt.tight_layout()
    plt.show()
```

```python
df = pd.read_csv('data/cache/AAPL.csv')
```

```python
explore_data(df)
```
